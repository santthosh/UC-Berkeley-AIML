### Will a Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaraunt near where you are driving. Would you accept that coupon and take a short detour to the restaraunt? Would you accept the coupon but use it on a sunbsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaraunt? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \\$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\\$20 - \\$50). 

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece. 





### Data Description

The attributes of this data set include:
1. User attributes
    -  Gender: male, female
    -  Age: below 21, 21 to 25, 26 to 30, etc.
    -  Marital Status: single, married partner, unmarried partner, or widowed
    -  Number of children: 0, 1, or more than 1
    -  Education: high school, bachelors degree, associates degree, or graduate degree
    -  Occupation: architecture & engineering, business & financial, etc.
    -  Annual income: less than \\$12500, \\$12500 - \\$24999, \\$25000 - \\$37499, etc.
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she buys takeaway food: 0, less than 1, 1 to 3, 4 to 8 or greater
    than 8
    -  Number of times that he/she goes to a coffee house: 0, less than 1, 1 to 3, 4 to 8 or
    greater than 8
    -  Number of times that he/she eats at a restaurant with average expense less than \\$20 per
    person: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    

2. Contextual attributes
    - Driving destination: home, work, or no urgent destination
    - Location of user, coupon and destination: we provide a map to show the geographical
    location of the user, destination, and the venue, and we mark the distance between each
    two places with time of driving. The user can see whether the venue is in the same
    direction as the destination.
    - Weather: sunny, rainy, or snowy
    - Temperature: 30F, 55F, or 80F
    - Time: 10AM, 2PM, or 6PM
    - Passenger: alone, partner, kid(s), or friend(s)


3. Coupon attributes
    - time before it expires: 2 hours or one day

**Investigating the Bar Coupons**

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


```python
# Creating a new DataFrame just for bar coupons

bar_coupon_data = data[data['coupon'] == 'Bar'] 
bar_coupon_data.head()
```

```python
bar_coupon_data.info()
```

```python
bar_coupon_data['Bar'].unique()
```

2. What proportion of bar coupons were accepted?


```python
# Total repsonses that said Yes divided by the total rows in the data frame 
bar_coupon_accepted_percentage = bar_coupon_data['Y'].sum()/len(bar_coupon_data) * 100
bar_coupon_accepted_percentage
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
# Lets first filter out the data frame to pick only those who went to a bar 3 or fewer times a month
bar_visitors_3_or_less = bar_coupon_data[bar_coupon_data['Bar'].isin(['never','less1','1~3',-1])]
bar_visitors_3_or_less['Bar'].unique()

# Now calculate their acceptance rate
bar_visitors_3_or_less_acceptance = bar_visitors_3_or_less['Y'].sum()/len(bar_visitors_3_or_less) * 100
bar_visitors_3_or_less_acceptance
```

```python
# Lets first filter out the data frame to pick only those who went to a bar 4 or more times a month
bar_visitors_4_or_more = bar_coupon_data[~bar_coupon_data['Bar'].isin(['never','less1','1~3', -1])]
bar_visitors_4_or_more['Bar'].unique()

# Now calculate their acceptance rate
bar_visitors_4_or_more_acceptance = bar_visitors_4_or_more['Y'].sum()/len(bar_visitors_4_or_more) * 100
bar_visitors_4_or_more_acceptance
```

```python
ratio_frequent_to_less_frequent_visitors = bar_visitors_4_or_more_acceptance / bar_visitors_3_or_less_acceptance
ratio_frequent_to_less_frequent_visitors
```

#### RESULT: Frequent Bar visitors who visited more than 3 times a month are twice more likely to accept the coupons than infrequent visitors who visited 3 or less times in a month

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
bar_coupon_data['age'].unique()
```

```python
# Lets first filter out the data frame to pick only those who went to a bar more than once a month and are over age 25
frequent_bar_visitors = bar_coupon_data[
    (~bar_coupon_data['Bar'].isin(['never','less1',-1])) & 
    (bar_coupon_data['age'].isin(['26','46','31','41','50plus','36']))]

frequent_bar_visitors['Bar'].unique()
```

```python
# construct a dataframe for everyother visitor
everyother_bar_visitors = bar_coupon_data[~bar_coupon_data.index.isin(frequent_bar_visitors.index)]

everyother_bar_visitors['Bar'].unique()
```

```python
# Lets calcuate acceptance for this frequent_bar_vistors
frequent_visitor_acceptance = frequent_bar_visitors['Y'].sum()/len(frequent_bar_visitors)*100
frequent_visitor_acceptance
```

```python
# Lets calcuate acceptance for this everyother_bar_vistor
everyother_bar_visitor_acceptance = everyother_bar_visitors['Y'].sum()/len(everyother_bar_visitors)*100
everyother_bar_visitor_acceptance
```

```python
ratio_frequent_to_everyone = frequent_visitor_acceptance / everyother_bar_visitor_acceptance
ratio_frequent_to_everyone
```

#### RESULT: This group of users visiting bar more than once a month and are of age group > 25 are 2 times more likely to accept the coupons than everyone else

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry. 


```python
bar_coupon_data['passanger'].unique()
```

```python
bar_coupon_data['occupation'].unique()
```

```python
# Let us call this candidate group as Group A
# Visits bar more than once a month
# Has passangers that are not a kid
# Occupation other than farming

group_a_df = bar_coupon_data[
    (~bar_coupon_data['Bar'].isin(['never','less1',-1])) & 
    (~bar_coupon_data['passanger'].isin(['Kid(s)'])) &
    (~bar_coupon_data['occupation'].isin(['Farming Fishing & Forestry']))]
group_a_df['Bar'].unique()

not_group_a_df = bar_coupon_data[~bar_coupon_data.index.isin(group_a_df.index)]

# lets calculate acceptance rate for each of these groups
group_a_acceptance = group_a_df['Y'].sum()/len(group_a_df)*100
group_a_acceptance

not_group_a_acceptance = not_group_a_df['Y'].sum()/len(not_group_a_df)*100
not_group_a_acceptance

# Compare them over 
ratio_group_a = group_a_acceptance / not_group_a_acceptance
ratio_group_a
```

#### RESULT: This group of users who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry are 2.4 times more likely to accept the coupons

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K. 



```python
bar_coupon_data['maritalStatus'].unique()
```

```python
bar_coupon_data['income'].unique()
```

```python
bar_coupon_data['RestaurantLessThan20'].unique()
```

```python
# Let us call this candidate group as Group B
# go to bars more than once a month, had passengers that were not a kid, and were not widowed OR
# go to bars more than once a month and are under the age of 30 OR
# go to cheap restaurants more than 4 times a month and income is less than 50K.

group_b_df = bar_coupon_data[
    ((~bar_coupon_data['Bar'].isin(['never','less1',-1])) & 
    (~bar_coupon_data['passanger'].isin(['Kid(s)'])) &
    (~bar_coupon_data['occupation'].isin(['Widowed']))) |
    
    ((~bar_coupon_data['Bar'].isin(['never','less1',-1])) & 
    bar_coupon_data['age'].isin(['26','21','below21'])) | 
     
     ((bar_coupon_data['RestaurantLessThan20'].isin(['4-8','gt8'])) & 
    bar_coupon_data['income'].isin(['Less than $12500','$12500 - $24999',
                                    '$25000 - $37499','$37500 - $49999']))
     ]

not_group_b_df = bar_coupon_data[~bar_coupon_data.index.isin(group_b_df.index)]

# lets calculate acceptance rate for each of these groups
group_b_acceptance = group_b_df['Y'].sum()/len(group_b_df)*100
group_b_acceptance

not_group_b_acceptance = not_group_b_df['Y'].sum()/len(not_group_b_df)*100
not_group_b_acceptance

# Compare them over 
ratio_group_b = group_b_acceptance / not_group_b_acceptance
ratio_group_b
```

#### RESULT: This group of users are also 2.4 times more likely to accept the coupons

```python
sns.countplot(data=coffee_coupon_data,x='passanger',hue='Y')
plt.title('BarPlot of Coffee Coupon By Passanger')
plt.show()
```

#### Most interesting observation in this bar plot is that alone drivers are more likely to accept the coupons followed by drivers with friends

```python
coffee_coupon_data_accepted = coffee_coupon_data[coffee_coupon_data['Y'] == 1]

sns.countplot(data=coffee_coupon_data_accepted,x='passanger',hue='temperature')
plt.title('BarPlot of Coffee Coupon By Passanger and Hue Temperature')
plt.show()
```

#### Coupon acceptance is much higher when the temperature is high around 80s

```python
# Based on the above plots lets pick the acceptance rate of drivers who are travelling with friends when the temperature is 80s
# Let us call this group - group c

group_c_df = coffee_coupon_data[
    ((coffee_coupon_data['temperature'] == 80) & 
    (coffee_coupon_data['passanger'].isin(['Friend(s)'])))]

not_group_c_df = coffee_coupon_data[~coffee_coupon_data.index.isin(group_c_df.index)]

# lets calculate acceptance rate for each of these groups
group_c_acceptance = group_c_df['Y'].sum()/len(group_c_df)*100
group_c_acceptance

not_group_c_acceptance = not_group_c_df['Y'].sum()/len(not_group_c_df)*100
not_group_c_acceptance

# Compare them over 
ratio_group_c = group_c_acceptance / not_group_c_acceptance
ratio_group_c
```

#### Based on the above results, passengers travelling with friends are 1.26 times more likely to accept coupons when the weather is in its 80s
